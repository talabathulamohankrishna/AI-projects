```python
#required packages

import pyttsx3
import speech_recognition as sr
import datetime
from datetime import datetime
from googlesearch import search
import wikipedia
import pywhatkit
import webbrowser
import os
import re
import sys
```

```python
import ctypes

# Define the LockWorkStation function from the Win32 API
def lock():
    ctypes.windll.user32.LockWorkStation()

```

```python
engine=pyttsx3.init()
voices=engine.getProperty('voices')
#print(voices[1])
engine.setProperty('voice',voices[1].id)

def respond(cmd):
    engine.say(cmd)
    engine.runAndWait()

def wish():
    respond('hi..')
    respond('what do u want?')



def take_cmd():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        print('listening...')
        r.pause_threshold=1
        audio=r.listen(source)
    try:
        speech=r.recognize_google(audio,language='en-in')
          
    except Exception as e:
        print('please say that again...')
        return 'None'
    return speech


if __name__=='__main__':
    wish()
    #for infinite loop
    while True:
        speech=take_cmd().lower()

    
        if 'date' in speech:
            today=datetime.today()
            year = today.year
            month = today.month
            day = today.day
            print("Today's date:", today)
            respond(day)
            respond(month)
        elif 'day' in speech:
            today=datetime.date.today()
            day_name=today.strftime('%A')
            respond('its..')
            respond(day_name)
            
        elif 'time' in speech:
            dt = datetime.now().time()

            # convert to 12 hour time format
            formatted_time = dt.strftime("%I:%M %p")
            respond(formatted_time)
            print(formatted_time) 

        elif 'what' or 'search' in speech:
            result=search(speech)
            print(result)

        elif 'weather' in speech:
            webbrowser.open("https://www.bing.com/search?pglt=43&q=whether+today&cvid=0e8cb479797d452b9fdf212e17d5dece&aqs=edge..69i57j0l8.6783j0j1&FORM=ANNTA1&PC=ASTS")

        elif 'are you there' in speech:
            respond('yes')
            respond('i am here')

        elif "lock" in speech:
            respond("locking the device")
            lock()
        
        elif 'link' in speech:
            num_results = 2
            for result in search(speech, num_results=num_results):
                print(result)
                respond(result)

        elif "where are you?" in speech:
            respond('at your service..')
    
        elif 'wikipedia' in speech:
            respond('one second...')
            speech=speech.replace('wikipedia','')
            results=wikipedia.summary(speech,sentences=2)
            respond(results)

        elif 'app' in speech:
            print('i will excecute')

        elif 'play' in speech:
            speech=speech.replace('play','')
            pywhatkit.playonyt(speech)
        elif 'open youtube' in speech:
            webbrowser.open('youtube.com')
        elif 'open google' in speech:
            webbrowser.open("google.com")

        elif "word" or 'doc' in speech:
            os.startfile("C:/ProgramData/Microsoft/Windows/Start Menu/Programs/Word.lnk")

        elif 'open excel' in speech:
            os.startfile("C:/ProgramData/Microsoft/Windows/Start Menu/Programs/Excel.lnk")

        elif 'play music' in speech:
            pywhatkit.playonyt('https://www.youtube.com/watch?v=ol3LPFfoIio&list=RDMMol3LPFfoIio&start_radio=1')

        elif 'vs code' in speech:
            os.startfile("C:/Users/ASUS/AppData/Roaming/Microsoft/Windows/Start Menu/Programs/Visual Studio Code")
        elif 'git' in speech:
            webbrowser.open('https://github.com/talabathulamohankrishna')
        elif 'add' or 'plus' or 'sum' in speech:
            numbers = re.findall(r'\d+', speech)
            numbers = [int(num) for num in numbers]
            num=0
            for i in numbers:
                num=i+num
            respond(num)
            
        elif 'subtract' or 'minus' or 'remove' in speech:
            numbers = re.findall(r'\d+', speech)
            numbers = [int(num) for num in numbers]
            sub=numbers[1]-numbers[0]
            respond(sub)
        elif 'goodbye' or 'bye' or 'good night' or 'pack up' or 'take care' in speech:
            respond("okk..")
            respond('see u')
            sys.exit()

    
```
