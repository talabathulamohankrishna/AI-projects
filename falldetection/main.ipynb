```python
import librosa
import os
import soundfile as sf
import numpy as np
import audiomentations as am

def augment_audio(audio_file, partition, n_augmentations):
  """Augments the audio file `n_augmentations` times and saves it to the augmented data directory."""

  # Load the audio file
  audio, sample_rate = librosa.load(audio_file)

  # Apply random augmentation techniques
  augmentations = am.Compose([
    am.AddBackgroundNoise(sounds_path=audio_file),
    am.PitchShift(),
    am.TimeStretch(),
    am.Gain(),
    am.Reverse(p=0.1),  # Reverse the audio
    am.ClippingDistortion(p=0.1),  # Simulate clipping distortion
    am.HighPassFilter(p=0.1),  # Apply high-pass filter
    am.LowPassFilter(p=0.1),  # Apply low-pass filter
    # Add more augmentation techniques as needed

  ])

  # Augment the audio file `n_augmentations` times
  for i in range(n_augmentations):
    augmented_audio = augmentations(audio, sample_rate=sample_rate)

    # Save the augmented audio file
    augmented_audio_file = os.path.join('aug_data', partition, 'audio', f'{audio_file.split(".")[0]}_{i}.mp3')
    sf.write(augmented_audio_file, augmented_audio, sample_rate)

folder_path="C:/Users/talab/Documents/GitHub/AI-projects/falldetection/data/train/nonfalls/"
# Get the list of all the mp3 files in the folder
audio_files = os.listdir(folder_path)

# Iterate over the audio files and augment them
for audio_file in audio_files:
  print(audio_file)
  if audio_file.endswith('.mp3'):
    audio_file = os.path.join(folder_path, audio_file)
    # Augment the audio file 30 times
    augment_audio(audio_file, "train", 25)
```

```python
import random
import os

def select_and_move_files(folder_path, destination_folder_path, n_files):
  """Selects `n_files` random files from the specified folder and moves them to the destination folder."""

  # Get a list of all the files in the folder
  files = os.listdir(folder_path)

  # Select `n_files` random files from the list
  selected_files = random.sample(files, n_files)

  # Move the selected files to the destination folder
  for file in selected_files:
    os.rename(os.path.join(folder_path, file), os.path.join(destination_folder_path, file))

# Example usage:

folder_path = "C:/Users/talab/Documents/GitHub/AI-projects/falldetection/data/train/nonfalls/"
destination_folder_path = "C:/Users/talab/Documents/GitHub/AI-projects/falldetection/data/test/nonfalls/"
n_files = 190

select_and_move_files(folder_path, destination_folder_path, n_files)
```

```python
def load_wav_16k_mono(filename):
    # Load encoded wav file
    file_contents = tf.io.read_file(filename)
    # Decode wav (tensors by channels) 
    wav, sample_rate = tf.audio.decode_wav(file_contents, desired_channels=1)
    # Removes trailing axis
    wav = tf.squeeze(wav, axis=-1)
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    # Goes from 44100Hz to 16000hz - amplitude of the audio signal
    wav = tfio.audio.resample(wav, rate_in=sample_rate, rate_out=16000)
    return wav
```

```python
def preprocess(file_path, label): 
    wav = load_wav_16k_mono(file_path)
    wav = wav[:48000]
    zero_padding = tf.zeros([48000] - tf.shape(wav), dtype=tf.float32)
    wav = tf.concat([zero_padding, wav],0)
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    return spectrogram, label
```

```python
filepath, label = positives.shuffle(buffer_size=10000).as_numpy_iterator().next()
spectrogram, label = preprocess(filepath, label)
```

```python
plt.figure(figsize=(30,20))
plt.imshow(tf.transpose(spectrogram)[0])
plt.show()
```

```python
data = data.map(preprocess)
data = data.cache()
data = data.shuffle(buffer_size=1000)
data = data.batch(16)
data = data.prefetch(8)
```

```python
train = data.take(36)
test = data.skip(36).take(15)
```

```python
samples, labels = train.as_numpy_iterator().next()
samples.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), activation='relu', input_shape=(1491, 257,1)))
model.add(Conv2D(16, (3,3), activation='relu'))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('Adam', loss='BinaryCrossentropy', metrics=[tf.keras.metrics.Recall(),tf.keras.metrics.Precision()])
```

```python
model.summary()
```

```python
hist = model.fit(train, epochs=4, validation_data=test)
```
